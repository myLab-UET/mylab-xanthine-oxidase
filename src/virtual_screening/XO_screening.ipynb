```python
import sys
sys.path.append("../../src/common")
import pandas as pd
import numpy as np
from pharmacy_common import PharmacyCommon
common = PharmacyCommon()
import sklearn.preprocessing as preprocessing
```

READ THE DATA

```python
train_test_validation_path = "../../data/train_test_validation_data/XO_train_test_validation_data.xlsx"
train_dataset = pd.read_excel(train_test_validation_path, sheet_name='train_dataset')

decoy_dataset_path = "../../data/decoy_data/decoy_dataset.xlsx"
decoy_dataset = pd.read_excel(decoy_dataset_path)
print("Decoy dataset size: ", len(decoy_dataset))

screen_dataset_path = "../../results/average_distance/avg_distance_screening_dataset_ECFP4_1024bits.xlsx"
screen_dataset = pd.read_excel(screen_dataset_path)
print("Screening dataset size: ", len(screen_dataset))
```

Encoding labels and fingerprints

```python
X_Decoy_ECFP4_1024 = common.gen_ecfp4_fpts(decoy_dataset["SMILES"],1024)
X_Decoy_ECFP4_2048 = common.gen_ecfp4_fpts(decoy_dataset["SMILES"],2048)
X_Decoy_ECFP6_1024 = common.gen_ecfp6_fpts(decoy_dataset["SMILES"],1024)
X_Decoy_ECFP6_2048 = common.gen_ecfp6_fpts(decoy_dataset["SMILES"],2048)
X_Decoy_MACCS      = common.gen_maccs_fpts(decoy_dataset["SMILES"])
```

```python
X_Screening_ECFP4_1024 = common.gen_ecfp4_fpts(screen_dataset["SMILES"],1024)
X_Screening_ECFP4_2048 = common.gen_ecfp4_fpts(screen_dataset["SMILES"],2048)
X_Screening_ECFP6_1024 = common.gen_ecfp6_fpts(screen_dataset["SMILES"],1024)
X_Screening_ECFP6_2048 = common.gen_ecfp6_fpts(screen_dataset["SMILES"],2048)
X_Screening_MACCS      = common.gen_maccs_fpts(screen_dataset["SMILES"])
```

```python
y_Train = np.array(train_dataset['Type'])
print(y_Train[0:5])

#One-hot encoder
label_encoder = preprocessing.LabelEncoder()
y_Train = label_encoder.fit_transform(y_Train)
print(y_Train[0:5])
```

LOAD MODELS

```python
from xgboost import XGBClassifier
from joblib import load

xgb_maccs_167 = XGBClassifier()
xgb_maccs_167.load_model("../../results/models/XGB_maccs_167bits.json")

xgb_ecfp4_1024 = XGBClassifier()
xgb_ecfp4_1024.load_model("../../results/models/XGB_ecfp4_1024bits.json")

xgb_ecfp4_2048 = XGBClassifier()
xgb_ecfp4_2048.load_model("../../results/models/XGB_ecfp4_2048bits.json")

xgb_ecfp6_1024 = XGBClassifier()
xgb_ecfp6_1024.load_model("../../results/models/XGB_ecfp6_1024bits.json")

xgb_ecfp6_2048 = XGBClassifier()
xgb_ecfp6_2048.load_model("../../results/models/XGB_ecfp6_2048bits.json")
```

```python
def dataset_predict(model, model_name, X_fingerprint, label_encoder, dataset):
    dataset_prediction = pd.DataFrame()
    dataset_prediction = dataset.copy()

    y_pred = model.predict(X_fingerprint)
    y_pred_proba = model.predict_proba(X_fingerprint)

    y_pred = label_encoder.inverse_transform(y_pred)
    dataset_prediction.loc[:, 'Prediction'] = y_pred
    dataset_prediction.loc[:, 'Probality for Active'] = y_pred_proba[:, 0]
    dataset_prediction.loc[:, 'Probality for Inactive'] = y_pred_proba[:, 1]
    dataset_prediction.loc[:, 'Model name'] = model_name

    return dataset_prediction
```

Predicting on the Decoy dataset

```python
xgb_ecfp4_1024_decoy_prediction = dataset_predict(model=xgb_ecfp4_1024, model_name="XGB_ECFP4_1024", X_fingerprint=X_Decoy_ECFP4_1024, label_encoder=label_encoder, dataset=decoy_dataset)
xgb_ecfp4_2048_decoy_prediction = dataset_predict(model=xgb_ecfp4_2048, model_name="XGB_ECFP4_2048", X_fingerprint=X_Decoy_ECFP4_2048, label_encoder=label_encoder, dataset=decoy_dataset)
xgb_maccs_decoy_prediction      = dataset_predict(model=xgb_maccs_167 , model_name="XGB_MACCS"     , X_fingerprint=X_Decoy_MACCS     , label_encoder=label_encoder, dataset=decoy_dataset)
xgb_ecfp6_1024_decoy_prediction = dataset_predict(model=xgb_ecfp6_1024, model_name="XGB_ECFP6_1024", X_fingerprint=X_Decoy_ECFP6_1024, label_encoder=label_encoder, dataset=decoy_dataset)
xgb_ecfp6_2048_decoy_prediction = dataset_predict(model=xgb_ecfp6_2048, model_name="XGB_ECFP6_2048", X_fingerprint=X_Decoy_ECFP6_2048, label_encoder=label_encoder, dataset=decoy_dataset)
```

```python
xgb_ecfp4_1024_decoy_prediction[xgb_ecfp4_1024_decoy_prediction["Prediction"]=="active"]
```

Predicting on the screening dataset

```python
xgb_ecfp4_1024_screening_prediction = dataset_predict(model=xgb_ecfp4_1024, model_name="XGB_ECFP4_1024", X_fingerprint=X_Screening_ECFP4_1024, label_encoder=label_encoder, dataset=screen_dataset)
xgb_ecfp4_2048_screening_prediction = dataset_predict(model=xgb_ecfp4_2048, model_name="XGB_ECFP4_2048", X_fingerprint=X_Screening_ECFP4_2048, label_encoder=label_encoder, dataset=screen_dataset)
xgb_maccs_screening_prediction      = dataset_predict(model=xgb_maccs_167 , model_name="XGB_MACCS"     , X_fingerprint=X_Screening_MACCS     , label_encoder=label_encoder, dataset=screen_dataset)
xgb_ecfp6_1024_screening_prediction = dataset_predict(model=xgb_ecfp6_1024, model_name="XGB_ECFP6_1024", X_fingerprint=X_Screening_ECFP6_1024, label_encoder=label_encoder, dataset=screen_dataset)
xgb_ecfp6_2048_screening_prediction = dataset_predict(model=xgb_ecfp6_2048, model_name="XGB_ECFP6_2048", X_fingerprint=X_Screening_ECFP6_2048, label_encoder=label_encoder, dataset=screen_dataset)
```

```python
xgb_ecfp4_1024_screening_prediction[xgb_ecfp4_1024_screening_prediction["Prediction"]=="active"]
```

Write to file

```python
output_path = "../../results/decoy_results/Decoy_prediction_results.xlsx"
with pd.ExcelWriter(output_path) as writer:
    xgb_ecfp4_1024_decoy_prediction.to_excel(writer, sheet_name="XGB-ECFP4-1024", index=False)
    xgb_ecfp4_2048_decoy_prediction.to_excel(writer, sheet_name="XGB-ECFP4-2048", index=False)
    xgb_ecfp6_1024_decoy_prediction.to_excel(writer, sheet_name="XGB-ECFP6-1024", index=False)
    xgb_ecfp6_2048_decoy_prediction.to_excel(writer, sheet_name="XGB-ECFP6-2048", index=False)
    xgb_maccs_decoy_prediction.to_excel(writer, sheet_name="XGB-MACCS", index=False)
```
